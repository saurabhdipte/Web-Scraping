```python
from autoscraper import AutoScraper
```

```python
amazon_url="https://www.amazon.in/s?k=iphones"
wanted_list=["Apple iPhone 14 (128 GB) - Midnight","₹71,999"]
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_pgko':'Title','rule_k5k4':'Price'})
scraper.keep_rules(['rule_pgko','rule_k5k4'])
scraper.save('Amazon-search')
```

```python
results=scraper.get_result_similar('https://www.amazon.in/s?k=samsung',group_by_alias=True)
```
